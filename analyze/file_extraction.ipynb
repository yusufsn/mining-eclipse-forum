# Import libraries

```python
import os, glob, csv, sys, re
from os import path
from urllib.parse import urlparse
import pandas as pd
from pprint import pprint
from bs4 import BeautifulSoup
from tld import get_tld
```

# Define directories

```python
userhome = os.path.expanduser('~')
dataset = userhome + r'/mining-forum/data/forumfiles1/'
analyze_dir = userhome + r'/mining-forum/analyze/'
```

# Counting number of files

```python
file_list = glob.iglob(dataset + '*')
filenum = ';'.join(file_list).split(';')
filenum.sort()
print ("Found " + str(len(filenum)) + " files")
```

# Get url from files

```python
url_data = []
mid_list = []
file_num = 0
for files in glob.iglob(dataset + "*"):
    file_num += 1
    sys.stdout.write('\rExtracting file number: %i' %(file_num) + ' / %i' %(len(filenum)))
    sys.stdout.flush()
    f = open(files)
    f = f.read()
    
    try:
        soup = BeautifulSoup(f, 'html.parser')
        topic = soup.find('a', attrs={'class':'MsgSubText'})
        for title in topic.stripped_strings:
            title = title
        
        msid = []
        soup = BeautifulSoup(f, 'html.parser')
        for message_id in soup.find_all('td', attrs={'class':'MsgR1 vt al MsgSubText'}):
            for mid in message_id.find_all('span', attrs={'class':'SmallText'}):
                for msgid in mid.find_all('a'):
                    msid.append(msgid.text)
        msid = msid[0]

        if msid not in mid_list:
            mid_list.append(msid)
            data = [file_num, msid, title]
            soup = BeautifulSoup(f, 'html.parser')
            for msg in soup.find_all('span', attrs={'class':'MsgBodyText'}):
                for atag in msg.find_all('a'):
                    ss = atag.text.replace(' ','')
                    parser = []
                    try:
                        link = re.search('^https?:\/\/.*[\r\n]*', ss)
                        link = link.group()
                        domain = urlparse(link)
                        tld_dt = get_tld(link)
                        parser = [link, domain.netloc, tld_dt]
                        data.insert(len(data),parser)
                    except:
                        pass
            n = 3
            link_name = []
            for i in range (0,len(data)-3):
                if data[n][0] not in link_name:
                    url_data.append([data[0], data[1], data[2], data[n][0], data[n][1], data[n][2]])
                    link_name.append(data[n][0])
                else:
                    pass
                n = n + 1
        else:
            pass
    except:
        pass

print ('\nFiles extraction are finished')
```

```python
with open (analyze_dir + r'csv_files/csv_forumfiles_01.csv', 'w') as csvfile:
    writers = csv.writer(csvfile, delimiter = ",")
    header = ['file_number', 'message_id', 'topic', 'url', 'domain', 'top_level_domain']
    writers.writerow(header)
    for item in url_data:
        writers.writerow(item)
print ('CSV file has been created')
```
