# Import libraries

```python
import os, csv
from os import path
import pandas as pd
```

# Define directories

```python
userhome = os.path.expanduser('~')
summ_dir = userhome + r'/mining-forum/analyze/csv_files/summarize/'
```

# Load links dataset

```python
links_data = pd.read_csv(summ_dir + r'all_joined_links_with_path.csv')
links_data = links_data[['domain']]
links_data
```

```python
# Rename "eclipse.org" to "www.eclipse.org" since both domain show the same page
links_data['domain'] = links_data['domain'].replace(['eclipse.org'], 'www.eclipse.org')
links_data
```

# Grouping the same domain and count them

```python
domain = links_data.groupby('domain')[['domain']].count()
domain_list = domain.sort_values('domain', ascending=False)
domain_list
```

```python
len(domain_list)
```

# Grouping link path

```python
df_link = pd.read_csv(summ_dir + r'all_joined_links_with_path.csv')
df_link
```

```python
df_path = df_link.groupby(['domain','link_path'])['link_path'].count().to_frame()
df_path = df_path.sort_values('link_path', ascending=False)
df_path
```

```python
df_path.to_csv(summ_dir + "rankings_of_linkpath.csv")
print ("CSV file has been created")
```
