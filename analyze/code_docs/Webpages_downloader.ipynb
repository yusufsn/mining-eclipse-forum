# Import libraries

```python
import urllib
```

```python
url = 'https://www.eclipse.org/forums/index.php/t/1000/'
```

```python
response = urllib.request.urlopen(url)
mystr = response.read()
webContent = mystr.decode("utf8")
response.close()

print (webContent)
```

```python
html_file = open("webfile","w")
html_file.write(webContent)
html_file.close()
```
