# Import libraries

```python
import os, csv
from os import path
import pandas as pd
```

# Define directories

```python
userhome = os.path.expanduser('~')
analyze_dir = userhome + r'/mining-forum/analyze/'
summ_dir = userhome + r'/mining-forum/analyze/csv_files/summarize/'
```

# Open thread dataset

```python
df_thread = pd.read_csv(summ_dir + 'list_of_thread_categories_and_users.csv')
df_thread
```

# Grouping the threads based on main topics

```python
cols = ['file_number','main_topic','subtopic','title','message_id','author_name','author_link','author_id','author_role']
```

```python
#Removing duplicate filenumber
df_main = df_thread[cols].drop_duplicates(subset=['file_number'], keep='first')
df_main
```

```python
df_main.to_csv(analyze_dir + "csv_files/datashare/all_final_threads.csv")
```

```python
df_maintopics = df_main[['main_topic']].groupby(df_main['main_topic']).agg(['count'])
df_maintopics
```

```python
df_maintopics.to_csv(summ_dir + 'list_of_maintopics_categories.csv')
print ("CSV file has been created")
```

# Grouping the threads based on subtopics

```python
df_subtopics = df_main.groupby(['main_topic','subtopic'])['subtopic'].count().to_frame()
df_subtopics.sort_values('subtopic', ascending=False)
```

```python
df_subtopics.to_csv(summ_dir + 'list_of_subtopics_categories.csv')
print ("CSV file has been created")
```

# Grouping number of user activities

```python
df_useractivity = df_main.groupby(['author_name', 'author_link', 'author_role'])['author_link'].count().to_frame()
df_useractivity.sort_values('author_link', ascending=False)
```

```python
df_useractivity.to_csv(summ_dir + 'list_of_user_distribution.csv')
print ("CSV file has been created")
```

# Grouping number of user roles

```python
df_userrole = df_main[['author_role']].groupby(df_main['author_role']).agg(['count'])
df_userrole
```

```python
df_userrole.to_csv(summ_dir + 'list_of_userrole_distribution.csv')
print ("CSV file has been created")
```
