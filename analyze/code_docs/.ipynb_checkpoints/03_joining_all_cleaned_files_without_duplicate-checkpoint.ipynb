# Import libraries

```python
import os, glob, csv, sys, shutil, math, random
from os import path
import pandas as pd
import numpy as np
```

# Define directories

```python
userhome = os.path.expanduser('~')
file_dir = userhome + r'/mining-forum/analyze/csv_files/clean_files/'
summ_dir = userhome + r'/mining-forum/analyze/csv_files/summarize/'
```

# Collect files

```python
files = []
for file in glob.iglob(file_dir + '*.csv'):
    files.append(file)
files.sort()
files
```

```python
len(files)
```

# Joining all csv files

```python
result = pd.DataFrame([])
for i, fname in enumerate(files):
    ff = pd.read_csv(fname, skiprows=0, usecols=[0,1,2,3])
    result = result.append(ff)
result.to_csv(summ_dir + 'all_joined_files_with_bot.csv', index=False)
print ("CSV file has been created")
```

```python
result = pd.read_csv(summ_dir + 'all_joined_files_with_bot.csv')
result = result.drop_duplicates(subset=['file_number'], keep='first')
```
